```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
G.add_node("Paper 1", type="paper")
G.add_node("Paper 2", type="paper")
G.add_node("Author 1", type="author")
G.add_node("Author 2", type="author")
G.add_node("Journal 1", type="journal")
G.add_node("Journal 2", type="journal")

# Add edges
G.add_edge("Paper 1", "Author 1", relationship="author")
G.add_edge("Paper 2", "Author 1", relationship="author")
G.add_edge("Paper 2", "Author 2", relationship="author")
G.add_edge("Paper 1", "Journal 1", relationship="published_in")
G.add_edge("Paper 2", "Journal 2", relationship="published_in")

# Add famous authors
G.add_node("Ada Lovelace", type="author")
G.add_node("Alan Turing", type="author")
G.add_edge("Paper 1", "Ada Lovelace", relationship="author")
G.add_edge("Paper 2", "Alan Turing", relationship="author")

# Add famous journals
G.add_node("Nature", type="journal")
G.add_node("Science", type="journal")
G.add_edge("Paper 1", "Nature", relationship="published_in")
G.add_edge("Paper 2", "Science", relationship="published_in")

# Extract subgraph of papers and authors
subgraph = G.subgraph([node for node, data in G.nodes(data=True) if data.get("type") in ["paper", "author"]])

# Plot the subgraph
pos = nx.spring_layout(subgraph)
plt.figure(figsize=(8, 6))
nx.draw_networkx(subgraph, pos=pos, with_labels=True, node_color="lightblue", node_size=500)
plt.title("Bibliographic Subgraph")
plt.axis("off")
plt.show()
```
